## **Distribution Analysis**

A method to describe the distribution of a variable and make inferences about population parameters

### **1.Histogram**

Description: A histogram visualizes the frequency of data points by dividing the range into intervals (bins).

Unique Feature: The bins parameter, which defines the granularity of the intervals.



```python
import matplotlib.pyplot as plt

# Histogram for price distribution
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=10, color='skyblue', edgecolor='black', label='Price Distribution')  # Unique: bins parameter
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(loc='upper right')  # Adding legend
plt.show()
```

### **2.KDE Plot(Kernel Density Estimate)**

Description: Visualizes the probability density function (PDF) of the dataset.

Unique Feature: The shade parameter fills the area under the curve.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# KDE Plot for price distribution
plt.figure(figsize=(10, 6))
sns.kdeplot(df['price'], fill=True, color='purple', label='Price Density')  # Unique: shade parameter
plt.title('KDE Plot of Price')
plt.xlabel('Price')
plt.ylabel('Density')
plt.legend(loc='upper right')  # Adding legend
plt.show()
```

### **3.Box Plot**

Description: Summarizes data distribution, highlighting median, quartiles, and outliers.

Unique Feature: Clearly visualizes outliers as dots.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Box Plot for price distribution
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'], color='orange', label='Price Summary')  # Unique: Highlights outliers
plt.title('Box Plot of Price')
plt.xlabel('Price')
plt.legend(['Price Distribution'], loc='upper right')  # Adding legend
plt.show()
```

### **4.Violin Plot**

Description: Combines KDE and box plot elements to show data variability.

Unique Feature: Visualizes both density and summary statistics.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Violin Plot for price distribution
plt.figure(figsize=(10, 6))
sns.violinplot(x=df['price'], color='teal', label='Price Variability')  # Unique: Combines KDE and box plot
plt.title('Violin Plot of Price')
plt.xlabel('Price')
plt.legend(['Price Distribution'], loc='upper right')  # Adding legend
plt.show()
```

## **Comparative Data**

These visualizations focus on comparing data across different categories or subcategories. They help highlight differences or trends in categorical data.



### **1.Bar Plot**

Description: Bar plots are used to compare discrete categories. They represent statistics (e.g., counts, means, or sums) for each category as a bar.

Unique Feature: Displays individual category statistics or counts, making it simple for direct comparisons.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar Plot for average price by furnishing status
plt.figure(figsize=(10, 6))
sns.barplot(x='furnishingstatus', y='price', data=df, palette='viridis', errorbar=None, label='Average Price')  # Unique: Statistical aggregate
plt.title('Bar Plot of Average Price by Furnishing Status')
plt.xlabel('Furnishing Status')
plt.ylabel('Average Price')
plt.legend(['Average Price'], loc='upper right')  # Adding legend
plt.show()
```

### **2.Grouped Bar Plot**

Description: A grouped bar plot shows subcategory comparisons within main categories, using separate bars for each subcategory.

Unique Feature: Provides detailed insights into subcategories within each group.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Grouped Bar Plot for price by furnishing status and air conditioning
plt.figure(figsize=(10, 6))
sns.barplot(x='furnishingstatus', y='price', hue='airconditioning', data=df, palette='Set2', ci=None)  # Unique: hue parameter for grouping
plt.title('Grouped Bar Plot: Price by Furnishing Status and Air Conditioning')
plt.xlabel('Furnishing Status')
plt.ylabel('Average Price')
plt.legend(title='Air Conditioning', loc='upper right')  # Adding legend
plt.show()
```

### **3.Stacked Bar Plot**

Description: A stacked bar plot displays cumulative distribution for each category, where subcategories are stacked atop one another.

Unique Feature: Shows both overall totals and the contribution of subcategories to each category.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Preparing data for stacked bar plot
stacked_data = df.groupby(['furnishingstatus', 'airconditioning'])['price'].mean().unstack()

# Stacked Bar Plot
plt.figure(figsize=(10, 6))
stacked_data.plot(kind='bar', stacked=True, color=['skyblue', 'orange'], edgecolor='black')
plt.title('Stacked Bar Plot: Price by Furnishing Status and Air Conditioning')
plt.xlabel('Furnishing Status')
plt.ylabel('Average Price')
plt.legend(title='Air Conditioning', loc='upper right')  # Adding legend
plt.show()
```

## **Relationships Plots**

### **1.Scatter Plot**
Description: A scatter plot displays individual data points, showing how two continuous variables are related. Each point represents an observation with its x and y coordinates corresponding to values of two variables.

Use Case: Ideal for identifying relationships or correlations between two numerical variables, such as price vs. area or age vs. income.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter Plot for price vs area
plt.figure(figsize=(10, 6))
sns.scatterplot(x='area', y='price', data=df, color='green', label='Price vs Area')  # Unique: Displays individual data points
plt.title('Scatter Plot: Price vs Area')
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend(loc='upper right')  # Adding legend
plt.show()
```

## **Part-to-Whole Relationships Plots**

These plots visualize how individual components contribute to the whole. They are great for showing proportions and percentages within a dataset. 
These visualizations are particularly useful when dealing with categorical data and are effective for summarizing the composition of data.



### **1. Pie Chart**

Description: A pie chart shows the proportions of different categories in a dataset. Each slice represents a category, and the size of the slice corresponds to the proportion of that category in relation to the whole.

Use Case: Ideal for visualizing simple proportions, such as market share by company or distribution of categories in a dataset.

```python
import matplotlib.pyplot as plt

# Pie Chart for the 'furnishingstatus' column
furnishing_counts = df['furnishingstatus'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(furnishing_counts, labels=furnishing_counts.index, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen'])
plt.title('Pie Chart: Furnishing Status Proportions')
plt.show()
```

### **2. Donut Chart**

Description: A donut chart is a variation of the pie chart where the center is left blank, allowing for additional information like the total value or some other metric to be displayed.

Use Case: Like a pie chart but more visually appealing and space-efficient when you want to display more information at the center, such as a title or total sum.

```python
# Donut Chart for the 'furnishingstatus' column
plt.figure(figsize=(8, 8))
plt.pie(furnishing_counts, labels=furnishing_counts.index, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen'], wedgeprops=dict(width=0.3))
plt.title('Donut Chart: Furnishing Status Proportions')
plt.show()
```

### **3.Treemap**

Description: A treemap visualizes hierarchical data using nested rectangles. Each branch of the hierarchy is represented by a colored rectangle, with the size of each rectangle proportional to the value it represents.

Use Case: Used for visualizing part-to-whole relationships in hierarchical structures. For example, showing sales distribution by region, category, and subcategory.

```python
import plotly.express as px

# Assuming 'df' is your DataFrame with columns like 'furnishingstatus' and 'prefarea'
fig = px.treemap(df, path=['prefarea', 'furnishingstatus'], values='price', color='furnishingstatus', hover_data=['price'])

fig.update_layout(title='Treemap: Price Distribution by Prefarea and Furnishing Status')
fig.show()
```

## **Categorical Data Plots**
Categorical data plots are used to visualize and compare data that falls into distinct categories or groups. These plots help to show the frequency, distribution, and relationships of categorical variables, making it easier to interpret patterns in the data.

Categorical data refers to variables that represent different categories or groups. These variables can be either:
Nominal: Categories with no natural order (e.g., gender, country).
Ordinal: Categories with a specific order (e.g., rating scales like "low", "medium", "high").

### **1.Strip Plot**

Description: A strip plot is used to display individual data points along a single axis, grouped by a categorical variable. It's useful for visualizing the distribution of a variable for different categories, showing the spread and density of data points.

Unique Feature: The strip plot shows all data points as individual marks, making it suitable for small datasets or when precise values for each data point need to be shown.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create Strip Plot for price vs furnishingstatus
sns.stripplot(x='furnishingstatus', y='price', data=df, jitter=True, palette='Set2')

plt.title('Strip Plot: Price vs Furnishing Status')
plt.show()
```

### **2. Swarm Plot**

Description: A swarm plot is an enhanced version of the strip plot. It adjusts the data points' positions to prevent overlap, improving visibility when there are multiple points in the same category.

Unique Feature: Unlike a strip plot, which randomly jitters points, a swarm plot arranges points such that they do not overlap, making it ideal for large datasets with many identical values.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create Swarm Plot for price vs furnishingstatus with hue parameter for color
sns.swarmplot(x='furnishingstatus', y='price', data=df, hue='furnishingstatus', palette='Set2', legend=False)

plt.title('Swarm Plot: Price vs Furnishing Status')
plt.show()
```

## **Statistical Summaries**

Statistical summaries are a collection of methods used to describe and analyze the central tendency, variability, and relationships within a dataset. These summaries give a high-level overview of key data patterns and statistical insights, which can help in drawing conclusions from raw data. Statistical summaries include metrics like mean, median, variance, and correlation, as well as visual representations like heatmaps and pair plots.

In the context of visualization, Heatmaps and Pair Plots are useful for quickly identifying relationships and patterns in numerical data.

### **1. Heatmap**

A heatmap is a matrix-like visualization used to represent data in a two-dimensional grid where individual values are represented as colors. It's particularly useful for visualizing correlation matrices, where color intensity reflects the strength of the relationship between variables.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
# Apply label encoding for categorical columns (all 'object' type columns)
label_encoder = LabelEncoder()

# Identify categorical columns that need to be encoded
categorical_columns = df.select_dtypes(include=['object']).columns

# Encoding categorical columns
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])



# Correlation matrix for the DataFrame after encoding
corr = df.corr()

# Create a heatmap to visualize correlations
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap: Correlation Matrix (With Categorical Encoded)')
plt.show()
```

### **2.Pair Plot**

A pair plot (or scatterplot matrix) creates scatter plots for every possible pair of features in a dataset, which helps in identifying relationships or correlations between pairs of features. It's useful when you want to explore pairwise relationships across multiple features at once.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pair plot for all numeric features in the DataFrame
sns.pairplot(df)

plt.title('Pair Plot: Pairwise Relationships of Features')
plt.show()
```

### **2.Choropleth Map Example (Visualizing Property Data by Region)**

If your dataset contains categorical data related to regions or countries, you can visualize data using a Choropleth Map. For example, if you have price data aggregated by regions or states, a choropleth map can represent the distribution of property prices across regions.

```python
#!pip install folium

import folium
import pandas as pd

# Sample data: Price by region (replace with your dataset)
data = {
    'Region': ['North', 'South', 'East', 'West'],
    'Price': [300000, 200000, 400000, 500000],  # Replace with actual aggregated data
    'Latitude': [25.7617, 34.0522, 41.8781, 36.1699],  # Sample latitudes for regions
    'Longitude': [-80.1918, -118.2437, -87.6298, -115.1398]  # Sample longitudes for regions
}

dfc = pd.DataFrame(data)

# Create a base map centered at an average latitude and longitude
m = folium.Map(location=[dfc['Latitude'].mean(), dfc['Longitude'].mean()], zoom_start=4)

# Add regions with choropleth map styling based on price
for idx, row in dfc.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=10,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f"Region: {row['Region']}<br>Price: ${row['Price']}"
    ).add_to(m)

# Display the map
m.save("choropleth_map.html")

#creates and saves a html file with the map
```

## **Hierarchical and Multivariate Data Visualizations**

To visualize hierarchical and multivariate data, you can use plots like Sunburst Plot and Parallel Coordinates Plot. These visualizations allow for exploring complex relationships in hierarchical structures and multiple variables.



### **1.Sunburst Plot (Hierarchical Data)**
A Sunburst Plot is a radial visualization that is especially useful for displaying hierarchical relationships. It is often used to represent how different categories are related to each other, showing parent-child relationships in a circular layout.

Example: Visualizing a Hierarchical Data Structure

```python
import pandas as pd
import plotly.express as px



# Convert categorical columns to string type (if not already)
df['furnishingstatus'] = df['furnishingstatus'].astype(str)
df['mainroad'] = df['mainroad'].astype(str)

# Create Sunburst Plot: Furnishing Status vs Mainroad
fig = px.sunburst(df,
                  path=['furnishingstatus', 'mainroad'],
                  values='price',  # Size of segments will be based on price
                  title="Sunburst Plot: Furnishing Status vs Mainroad")
fig.show()
```

### **2.Parallel Coordinates Plot**
 A powerful tool for visualizing multivariate data (i.e., data with many variables or features). In a parallel coordinates plot, each feature or variable is represented as a vertical axis. Data points are represented by lines connecting their values on each axis, allowing you to see relationships between multiple variables at once.

```python
import plotly.express as px



# Encoding categorical columns to numeric values (if not already encoded)
df['furnishingstatus'] = df['furnishingstatus'].map({'Furnished': 1, 'Unfurnished': 0, 'Semi-Furnished': 0.5})
df['mainroad'] = df['mainroad'].map({'Yes': 1, 'No': 0})
df['guestroom'] = df['guestroom'].map({'Yes': 1, 'No': 0})
df['basement'] = df['basement'].map({'Yes': 1, 'No': 0})
df['hotwaterheating'] = df['hotwaterheating'].map({'Yes': 1, 'No': 0})
df['airconditioning'] = df['airconditioning'].map({'Yes': 1, 'No': 0})
df['parking'] = df['parking'].astype(int)  # Ensuring it's an integer

# Create Parallel Coordinates Plot: Price, Area, Bedrooms, Bathrooms, and Furnishing Status
fig = px.parallel_coordinates(df,
                              dimensions=['price', 'area', 'bedrooms', 'bathrooms', 'furnishingstatus'],
                              color='price', 
                              color_continuous_scale='Viridis',
                              title="Parallel Coordinates Plot: Property Features")
fig.show()
```

## **Specialty and Advanced Visualization**

These visualizations are often used for advanced data exploration when you have multidimensional data or want to represent complex relationships. Heres a breakdown of each type:

### **1.3D Scatter Plot: Show Relationships in Three Dimensions**
Description:
A 3D scatter plot is a visualization tool that displays three-dimensional data points. It allows you to analyze relationships among three continuous variables by positioning data points in three-dimensional space (x, y, z axes).

Use Cases:
Exploring multivariate data: To understand the relationship between three continuous features in your dataset.
Real Estate: Understanding how price, area, and bedrooms interact in a three-dimensional space.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x = df['area']
y = df['bedrooms']
z = df['price']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z)
ax.set_xlabel('Area')
ax.set_ylabel('Bedrooms')
ax.set_zlabel('Price')

plt.title('3D Scatter Plot: Area, Bedrooms, and Price')
plt.show()
```

### **2.Radar Chart: Compare Multivariate Data Over a Radial Grid**
Description:
A radar chart (also called a spider chart) represents multivariate data with a set of values plotted on axes that radiate from a central point. Each axis represents one variable, and the values are plotted as a series of connected lines, forming a polygon.

Use Cases:
Comparing features: Comparing the strengths of different categories across several features.
Performance Analysis: Evaluating multiple variables like sales, marketing effectiveness, customer satisfaction, etc.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
labels = ['Area', 'Bedrooms', 'Bathrooms', 'Parking', 'Price']
values = [2000, 3, 2, 1, 150000]

# Create a radar chart
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
values += values[:1]  # Closing the loop
angles += angles[:1]  # Closing the loop

fig, ax = plt.subplots(figsize=(6, 6), dpi=120, subplot_kw=dict(polar=True))
ax.plot(angles, values, linewidth=2, linestyle='solid')
ax.fill(angles, values, alpha=0.3)

ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

plt.title('Radar Chart: Property Features')
plt.show()
```

### **3.Stream Plot: Represent Flow Data (e.g., Vector Fields)**
Description:
A stream plot is used to visualize vector fields or flow data. It shows the movement or flow of quantities, such as air or water currents, represented as arrows or lines.

Use Cases:
Fluid Dynamics: Representing flow patterns of fluids, such as wind or ocean currents.
Data Science: Flow of data in a machine learning algorithm or network traffic analysis.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for stream plot
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2

plt.streamplot(X, Y, U, V, color='blue')
plt.title('Stream Plot: Vector Field')
plt.show()
```

### **4. Contour Plot: Show Contour Levels of a 3D Surface in 2D**
Description:
A contour plot is a two-dimensional representation of a three-dimensional surface. It shows contour lines or filled regions that represent levels of a variable on a 2D plane, corresponding to the value of a third variable in a 3D space.

Use Cases:
Topography: Representing the shape of land or surface features.
Data Science: Visualizing decision boundaries in machine learning or visualizing continuous fields like temperature or pressure.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for contour plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a contour plot
plt.contour(X, Y, Z, 20, cmap='viridis')
plt.title('Contour Plot: 3D Surface in 2D')
plt.colorbar()
plt.show()
```
